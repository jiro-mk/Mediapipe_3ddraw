```python
import csv
import pprint
import cv2
import k3d
import numpy as np


with open('landmark.csv') as f:
    reader = csv.reader(f)
    data = [row for row in reader]

    
#Choose Hand Landmarks
l1=[0,1,2,3,4]

l2=[0,5,6,7,8]

l3=[9,10,11,12]

l4=[13,14,15,16]

l5=[0,17,18,19,20]

l6=[5,9,13,17]


def draw_landmarks(x,y,z):
    xx=[a for a in x if a != '']
    yy=[a for a in y if a != '']
    zz=[a for a in z if a != '']
    x=np.array(xx).reshape(-1, 1)
    y=np.array(yy).reshape(-1, 1)
    z=np.array(zz).reshape(-1, 1)
    points =np.concatenate([x, y, z], 1)

    points1=points[l1,:]

    points2=points[l2,:]

    points3=points[l3,:]

    points4=points[l4,:]

    points5=points[l5,:]

    points6=points[l6,:]

    #Add plot data
    plt_line1 = k3d.line(points1, shader='mesh', width=0.005)
    plt_line2 = k3d.line(points2, shader='mesh', width=0.005)
    plt_line3 = k3d.line(points3, shader='mesh', width=0.005)
    plt_line4 = k3d.line(points4, shader='mesh', width=0.005)
    plt_line5 = k3d.line(points5, shader='mesh', width=0.005)
    plt_line6 = k3d.line(points6, shader='mesh', width=0.005)
    plt_line = [plt_line1,plt_line2,plt_line3,plt_line4,plt_line5,plt_line6]
    
    return plt_line



plot = k3d.plot(name='points')
plot.display()
plt_line_old_R = []
plt_line_old_L = []

for i in range(60,len(data)):
    if data[i][0] != '' :
        x_R=np.array(data[i][0:62:3]).astype(float)
        y_R=np.array(data[i][1:62:3]).astype(float)
        z_R=np.array(data[i][2:63:3]).astype(float)
        plt_line_R = draw_landmarks(x_R,y_R,z_R)
        for r, r_o in zip(plt_line_R,plt_line_old_R):
            plot += r
            plot -= r_o
        plt_line_old_R = plt_line_R

    if data[i][63] != '' :
        x_L=np.array(data[i][63::3]).astype(float)+np.array([0.3])
        y_L=np.array(data[i][64::3]).astype(float)
        z_L=np.array(data[i][65::3]).astype(float) 
        plt_line_L = draw_landmarks(x_L,y_L,z_L)
        for l, l_o in zip(plt_line_L,plt_line_old_L):
            plot += l
            plot -= l_o
        plt_line_old_L = plt_line_L

    
```
